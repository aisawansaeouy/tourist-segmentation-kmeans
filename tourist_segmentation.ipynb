```python
# การจัดการข้อมูลทางการเงิน
spending_cols = [
    '1) ค่าอาหาร จำนวนกี่บาท',
    '2) ค่าพาหนะ/น้ำมัน จำนวนกี่บาท',
    '3) ค่าที่พัก จำนวนกี่บาท',
    '4) ค่าซื้อสินค้าภายในงาน  จำนวนกี่บาท',
    '5) ค่าซื้อบัตร/สลากกาชาด จำนวนกี่บาท',
    '6) ค่าใช้จ่ายอื่น ๆ  จำนวนกี่บาท'
]

def clean_currency(x):
  if isinstance(x, str):
    clean_str = ''.join(c for c in x if c.isdigit() or c == '.')
    return float(clean_str) if clean_str else 0.0
  return float(x) if isinstance(x, (int, float)) else 0.0

for col in spending_cols:
  df[col] = df[col].apply(clean_currency)

df['Total_Spending'] = df[spending_cols].sum(axis=1)

df.Total_Spending
```

```python
# การจัดการความพึงพอใจด้านกิจกรรม (2.10)
sat_activity_cols = [c for c in df.columns if '2.10' in c]

for col in sat_activity_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    mean_val = df[col].mean()
    df[col] = df[col].fillna(mean_val)

df['Sat_Activities_Score'] = df[sat_activity_cols].mean(axis=1)

df.Sat_Activities_Score
```

```python
# การจัดการความพึงพอใจสิ่งอำนวยความสะดวก (2.11)
sat_facility_cols = [c for c in df.columns if '2.11' in c]

for col in sat_facility_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Sat_Facilities_Score'] = df[sat_facility_cols].mean(axis=1)

df.Sat_Facilities_Score
```

```python
# การคัดกรองข้อมูล
df_clean = df.dropna(subset=sat_facility_cols).copy()

print(f"จำนวนข้อมูลที่นำมาวิเคราะห์: {len(df_clean)}")
```

# Clustering K-Means

## Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df_clean[['Total_Spending', 'Sat_Activities_Score', 'Sat_Facilities_Score']]

# ปรับสเกล
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# รัน Loop หาค่า K
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# วาดกราฟ
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.title('Elbow Method (3 Factors: Money, Activity, Facility)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

## K-Mean (k=4)

```python
from sklearn.cluster import KMeans

# รัน model k=4
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# บันทึกกลุ่มของแต่ละข้อมูล
df_clean['Cluster_ID'] = clusters
```

```python
# สร้างตัวแปรขาประจำ (Revisit_Label)
target_col = ' 2.1 ท่านเคยมาชมงาน เทิดพระเกียรติฯ และงานกาชาดจังหวัดชุมพร จำนวนกี่ครั้ง (รวมครั้งนี้ด้วย)'
df_clean['Revisit_Label'] = df_clean[target_col].apply(lambda x: 0 if '1 - 2' in str(x) else 1)

# สรุปผล
summary = df_clean.groupby('Cluster_ID')[['Total_Spending', 'Sat_Activities_Score', 'Sat_Facilities_Score']].mean()
summary['Count'] = df_clean['Cluster_ID'].value_counts()
summary['Percentage'] = (summary['Count'] / len(df_clean)) * 100
summary['Loyal_Rate (%)'] = df_clean.groupby('Cluster_ID')['Revisit_Label'].mean() * 100

# คำนวณค่า "รวม" (Total) ของทั้ง 695 คน
total_row = pd.DataFrame({
    'Total_Spending': [df_clean['Total_Spending'].mean()],
    'Sat_Activities_Score': [df_clean['Sat_Activities_Score'].mean()],
    'Sat_Facilities_Score': [df_clean['Sat_Facilities_Score'].mean()],
    'Count': [len(df_clean)],
    'Percentage': [100.00],
    'Loyal_Rate (%)': [df_clean['Revisit_Label'].mean() * 100]
}, index=['รวม'])

# เอามาต่อกัน (Concat)
summary_final = pd.concat([summary, total_row])

# จัดลำดับคอลัมน์ให้สวยงาม
cols_order = ['Count', 'Percentage', 'Total_Spending', 'Sat_Activities_Score', 'Sat_Facilities_Score', 'Loyal_Rate (%)']
summary_final = summary_final[cols_order]

print("\n" + "="*50)
print("สรุปผลการจำแนกกลุ่ม")
print("="*50)
print(summary_final.round(2))
```

```python
# --- 1. แยกชื่อคอลัมน์กิจกรรม (2.10) และ สิ่งอำนวยความสะดวก (2.11) ---
act_cols_detail = [c for c in df_clean.columns if '2.10' in c]
fac_cols_detail = [c for c in df_clean.columns if '2.11' in c]

# ==============================================================================
# 1. ตารางเจาะลึกด้านสิ่งอำนวยความสะดวก (Facilities)
# ==============================================================================
# 1.1 หาค่าเฉลี่ยรายกลุ่ม (Cluster) ของแต่ละหัวข้อ
fac_table = df_clean.groupby('Cluster_ID')[fac_cols_detail].mean().T

# 1.2 เพิ่มบรรทัด "ค่าเฉลี่ยรวมด้านสิ่งอำนวยฯ" ข้างล่างสุด
# (เอาคะแนน Sat_Facilities_Score มาหาค่าเฉลี่ยรายกลุ่ม)
fac_score_means = df_clean.groupby('Cluster_ID')['Sat_Facilities_Score'].mean()

# เพิ่มเข้าไปเป็นแถวสุดท้าย
fac_table.loc['ค่าเฉลี่ยรวมด้านสิ่งอำนวยฯ'] = fac_score_means

print("\n" + "="*50)
print("ตารางเจาะลึกความพึงพอใจด้านสิ่งอำนวยความสะดวก (Facilities)")
print("="*50)
print(fac_table.round(2))


# ==============================================================================
# 2. ตารางเจาะลึกด้านกิจกรรม (Activities)
# ==============================================================================
# 2.1 หาค่าเฉลี่ยรายกลุ่ม (Cluster) ของแต่ละหัวข้อ
act_table = df_clean.groupby('Cluster_ID')[act_cols_detail].mean().T

# 2.2 เพิ่มบรรทัด "ค่าเฉลี่ยรวมด้านกิจกรรม" ข้างล่างสุด
# (เอาคะแนน Sat_Activities_Score มาหาค่าเฉลี่ยรายกลุ่ม)
act_score_means = df_clean.groupby('Cluster_ID')['Sat_Activities_Score'].mean()

# เพิ่มเข้าไปเป็นแถวสุดท้าย
act_table.loc['ค่าเฉลี่ยรวมด้านกิจกรรม'] = act_score_means

print("\n" + "="*50)
print("ตารางเจาะลึกความพึงพอใจด้านกิจกรรม (Activities)")
print("="*50)
print(act_table.round(2))
```

# Logistic Regression

```python
import statsmodels.api as sm

# ตัวแปรอิสระที่ใช้ (อ้างอิง Kotler & Keller, Crompton)
var_mapping = {
        '1.1 เพศ': 'Gender',
        '1.2 อายุ': 'Age',
        '2.6 ท่านเดินทางมาเที่ยวชมงานในลักษณะใด': 'Companion'
}
```

```python
for i in range(4):

    print("\n" + "-" * 60)
    print(f"[ วิเคราะห์กลุ่มที่ {i} ]")
    print("-" * 60)

    # 1) ดึงข้อมูลเฉพาะกลุ่มนั้นออกมา
    sub_df = df_clean[df_clean['Cluster_ID'] == i].copy()

    # 2) เช็คจำนวนคน (Safety Check)
    n_samples   = len(sub_df)
    n_revisit   = sub_df['Revisit_Label'].sum()  # ขาประจำ (1)
    n_norevisit = n_samples - n_revisit          # ทั่วไป (0)

    print(f"ข้อมูล: ทั้งหมด {n_samples} คน (ขาประจำ {n_revisit} | ทั่วไป {n_norevisit})")

    # ถ้าคนน้อยเกินไป หรือไม่มีใครขาประจำ -> ข้าม
    if (n_samples < 30) or (n_revisit < 5) or (n_norevisit < 5):
        print("ข้ามการวิเคราะห์: ข้อมูลไม่เพียงพอ (N < 30 หรือ Events < 5)")
        continue

    try:
        # ----------------------------------------------------
        # 3) เตรียมตัวแปรอิสระ (X)
        # ----------------------------------------------------
        found_cols = []
        rename_dict = {}

        for partial_name, new_eng_name in var_mapping.items():
            matches = [c for c in df_clean.columns if partial_name in c]
            if matches:
                real_col_name = matches[0]
                found_cols.append(real_col_name)
                rename_dict[real_col_name] = new_eng_name

        X_reg = sub_df[found_cols].copy()
        X_reg.rename(columns=rename_dict, inplace=True)

        # แปลงเป็น Dummy Variables
        X_dummy = pd.get_dummies(X_reg, drop_first=True).astype(float)

        # ----------------------------------------------------
        # 4) เตรียมตัวแปรตาม (Y) และรันโมเดล
        # ----------------------------------------------------
        Y_target = sub_df['Revisit_Label']

        X_with_const = sm.add_constant(X_dummy)

        model  = sm.Logit(Y_target, X_with_const)
        result = model.fit(disp=0)

        # ----------------------------------------------------
        # 5) ดึงค่าสถิติ
        # ----------------------------------------------------
        params = result.params
        se = result.bse
        wald = result.tvalues ** 2
        pvals = result.pvalues
        exp_b = np.exp(params)

        stats_table = pd.DataFrame({
            'B (Coef)': params,
            'S.E.': se,
            'Wald': wald,
            'Sig.': pvals,
            'Exp(B)': exp_b
        })

        # ----------------------------------------------------
        # 6) แสดงผลลัพธ์
        # ----------------------------------------------------
        prsquared = result.prsquared
        print(f"\nPseudo R-squared = {prsquared:.4f}")

        print("-" * 95)
        print(f"{'Variable':<30} | {'B':<8} | {'S.E.':<8} | {'Wald':<8} | {'Sig.':<8} | {'Exp(B)':<8}")
        print("-" * 95)

        found_sig = False
        for col in stats_table.index:

            sig_val = stats_table.loc[col, 'Sig.']

            if col == 'const' or sig_val < 0.10:
                found_sig = True

                star = ""
                if sig_val < 0.05:
                    star = "*"
                elif sig_val < 0.10:
                    star = "**"

                b_val     = stats_table.loc[col, 'B (Coef)']
                se_val    = stats_table.loc[col, 'S.E.']
                wald_val  = stats_table.loc[col, 'Wald']
                exp_b_val = stats_table.loc[col, 'Exp(B)']

                print(f"{col:<30} | {b_val:.4f} | {se_val:.4f} | {wald_val:.4f} | {sig_val:.4f}{star:<2} | {exp_b_val:.4f}")

        if not found_sig:
            print("ไม่พบตัวแปรที่มีนัยสำคัญทางสถิติ (Sig < 0.10)")

        print("-" * 95)

    except Exception as e:
        print(f"Error ในกลุ่ม {i}: {e}")
```

```python
# ส่วนตรวจสอบเพิ่มเติม

print("\n" + "=" * 80)
print("EXCLUSIVE CHECK: วิเคราะห์เพิ่มเติมสำหรับกลุ่ม 0 และกลุ่ม 2")
print("=" * 80)

# กำหนดตัวแปรที่ต้องการตรวจสอบ
check_vars = {
    'เพศ': '1.1 เพศ',
    'อายุ': '1.2 อายุ',
    'เพื่อนร่วมทาง': '2.6 ท่านเดินทางมาเที่ยวชมงานในลักษณะใด'
}

# ------------------------------------------------------------------------------
# 1) วิเคราะห์กลุ่มที่ 0 (ตรวจสอบสาเหตุของ Singular Matrix)
# ------------------------------------------------------------------------------
print("\n[ วิเคราะห์กลุ่มที่ 0 ] : ตรวจสอบ Perfect Separation")
group0 = df_clean[df_clean['Cluster_ID'] == 0]

print(f"จำนวนคนทั้งหมด: {len(group0)} คน")
print(f"สัดส่วนตัวแปรตาม: {group0['Revisit_Label'].value_counts().to_dict()} (0=ทั่วไป, 1=ขาประจำ)")

print("\nผลการตรวจสอบ Cross-tab (หากพบค่า 0 อาจเป็นสาเหตุของ Error):")

found_culprit = False

for label, col_name in check_vars.items():
    real_col = [c for c in df_clean.columns if col_name in c][0]

    ct = pd.crosstab(group0[real_col], group0['Revisit_Label'])

    has_zero = (ct == 0).any().any()
    status = "พบค่า 0" if has_zero else "ปกติ"

    if has_zero:
        found_culprit = True

    print(f"\n--- ตัวแปร: {label} --- สถานะ: {status}")
    print(ct)

if found_culprit:
    print("\nสรุป: พบตัวแปรที่อาจก่อให้เกิด Perfect Separation")

# ------------------------------------------------------------------------------
# 2) วิเคราะห์กลุ่มที่ 2 (ตรวจสอบความเป็นเนื้อเดียวกันของข้อมูล)
# ------------------------------------------------------------------------------
print("\n" + "-" * 80)
print("\n[ วิเคราะห์กลุ่มที่ 2 ] : ตรวจสอบความเหมือนกันของข้อมูล (Homogeneity)")
group2 = df_clean[df_clean['Cluster_ID'] == 2]

print(f"จำนวนคนทั้งหมด: {len(group2)} คน")

print("\nเปรียบเทียบร้อยละของการมาซ้ำ (Revisit = 1):")

for label, col_name in check_vars.items():
    real_col = [c for c in df_clean.columns if col_name in c][0]

    print(f"\n--- ตัวแปร: {label} ---")

    props = group2.groupby(real_col)['Revisit_Label'].mean() * 100

    print(props.map('{:.2f}%'.format))

    diff = props.max() - props.min()
    print(f"ความแตกต่างสูงสุดระหว่างกลุ่มย่อย: {diff:.2f}%")

    if diff < 10:
        print("ข้อสังเกต: ความแตกต่างต่ำ อาจไม่พบความมีนัยสำคัญทางสถิติ")

print("\n" + "=" * 80)
```
